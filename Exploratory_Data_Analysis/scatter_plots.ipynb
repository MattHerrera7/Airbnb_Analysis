```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
# Test CSV file
properties_df = pd.read_csv("csv/listings_all.csv")
properties_df.describe()
```

```python
# Tried 700, but that still too high.
properties_reduced_df = properties_df.loc[properties_df["night_price"] <= 600]

# Plot data and regression model fits across a FacetGrid.
# This function combines regplot() and FacetGrid. It is intended as a convenient 
# interface to fit regression models across conditional subsets of a dataset.
sns.lmplot(data=properties_reduced_df, x="star_rating", y="night_price", fit_reg=True, hue="airbnb_city")
```

```python
# Regression Model
sns.relplot(data=properties_reduced_df, x="star_rating", y="night_price", style="airbnb_city", hue="airbnb_city")
```

```python
sns.relplot(data=properties_reduced_df, x="star_rating", y="night_price",  style="airbnb_city", hue="airbnb_city")
```
